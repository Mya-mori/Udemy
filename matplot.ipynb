Matplotlib

```
import numpy as np
import matplotlib.pyplot  as plt
```

```
x = np.linspace(0, 10, 10)
y = x**1.5
plt.plot(x, y, "-r")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show()
```

```
plt.subplot(1, 2, 1)
plt.plot(x, y, "r") #red

plt.subplot(1, 2, 2)
plt.plot(x, y, "b") #blue
```

```
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel("x label")
axes.set_ylabel("y label")
axes.set_title("label")

plt.show()
```

```
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
#                        位置 横   縦   
#                                 大きさ  横　縦

axes1.plot(x, y)
axes1.set_title("Large one")

axes2.plot(y, x)
axes2.set_title("Small one")

plt.show()
```

```
fig, axes = plt.subplots(nrows = 3, ncols = 3)
```

```
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.show()
```

```
fig = plt.figure(figsize = (4, 4))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, label = "1001")
ax.plot(y, x, label = "0110")

ax.legend(loc = 10)
```

```
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = "red", linewidth = 1, alpha =10, linestyle = "dashed", marker = "o", markersize =10,
        markerfacecolor = "yellow") 
#color --- line color
#linewidth(lw) --line width
#alpha
#linestyle(Is)
#marker
#markersize
#markerfacecolor

ax.set_xlim([0, 1])
ax.set_xlim([0, 2])
```
